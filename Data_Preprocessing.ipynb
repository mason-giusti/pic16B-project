```python
from PIL import Image
import os
```

```python

# Abstract Expressionism
# Define paths
input_dir = "/Users/bridgetbidwell/PIC_16B/WIKIART/Abstract_Expressionism"
output_dir = "/Users/bridgetbidwell/Projects/pic16B-project"
new_size = (224, 224)  # Change this to the target size, e.g., 224x224

```

```python
# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)
```

```python
# THIS WORKS, 

# Loop through each image in the input directory
for img_name in os.listdir(input_dir):
    img_path = os.path.join(input_dir, img_name)
    try:
        # Open and resize the image
        with Image.open(img_path) as img:
            resized_img = img.resize(new_size, Image.LANCZOS)
            
            # Save the resized image
            resized_img.save(os.path.join(output_dir, img_name))
            print(f"Resized and saved {img_name}")
    except Exception as e:
        print(f"Could not process {img_name}: {e}")
```
